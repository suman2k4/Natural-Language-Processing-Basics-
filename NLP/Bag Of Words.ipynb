```python
import pandas as pd
```

```python
ges = pd.read_csv('spam.csv', names=['label', 'message'], encoding='latin1')
```

```python
messages = ges['message'].values
```

```python
messages 
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus = []
for text in messages:  # Directly iterate through array elements
    review = re.sub('[^a-zA-Z]', ' ', str(text))  # Ensure text is string
    review = review.lower()
    review = ' '.join(review.split())  # Clean extra whitespace
    corpus.append(review)
```

```python
corpus
```

## Create Bag Of Words

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=100,binary=True)
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
X
```

### N-Grams

```python
cv.vocabulary_
```

```python
## Create the Bag OF Words model with ngram
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=100,binary=True,ngram_range=(2,3))
X=cv.fit_transform(corpus).toarray()
```
